```python
import pandas as pd

file_path = input("Enter the path to the csv file which has .csv extension: (without '')")
if not file_path.endswith('.csv'):
    raise ValueError("File extension must be '.csv'")

sales_data = pd.read_csv(file_path)

categories = sales_data['Category'].unique()

product_category = sales_data.groupby('Category')['Product'].unique()

sales_per_category = sales_data.groupby('Category')['Sales'].sum()
sales_per_product = sales_data.groupby('Product')['Sales'].sum()

sales_lst = []
for category in categories:
    for product in product_category[category]:
        sales_percentage = ((sales_per_product[product]/sales_per_category[category])*100).round(2)
        sales_lst.append([category,product,sales_percentage])

data = pd.DataFrame(sales_lst, columns = ['Category','Product','Sales_percentage'])

# Save the result to a new CSV file
output_filename = "Bhavesh_products.csv"
data.to_csv(output_filename, index=False)

print("Output saved to", output_filename)
```

```python
import pandas as pd

# Input file path and validate extension
file_path = input('Enter the path to the csv file which has .csv extension: ')
if not file_path.endswith('.csv'):
    raise ValueError("File extension must be '.csv'")

# Read CSV file into DataFrame
sales_data = pd.read_csv(file_path)

# Group by category and product
grouped = sales_data.groupby(['Category', 'Product'])

# Calculate total sales per category
sales_per_category = sales_data.groupby('Category')['Sales'].transform('sum')

# Calculate sales percentage for each product within its category
data = pd.DataFrame({
    'Category': sales_data['Category'],
    'Product': sales_data['Product'],
    'Sales_percentage': ((grouped['Sales'].transform('sum') / sales_per_category) * 100).round(2)
})

# Save the result to a new CSV file
output_filename = "Bhavesh_products.csv"
data.to_csv(output_filename, index=False)

print("Output saved to", output_filename)
```
