The scipy.stats module contains various functions for statistical calculations and tests.


## Binomial distribution

Approach :
<ul>
    <li>Define n and p.</li>
    <li>Define a list of values of r from 0 to n.</li>
    <li>Get mean and variance.</li>
    <li>For each r, calculate the pmf and store it in a list.</li>
</ul>list.

```python
from scipy.stats import binom
```

```python
#Setting the parameters
n=6
p=0.6

#Define the list of r values
r_values = list(range(n+1))

#btaining mean and variance
mean, var = binom.stats(n,p)
```
